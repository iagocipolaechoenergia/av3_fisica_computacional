```python
import numpy as np
import matplotlib.pyplot as plt

# === Parte (a)  Curva deltoide ===========================================
theta = np.linspace(0, 2*np.pi, 1000)          # 1000 pontos entre 0 e 2
x = 2*np.cos(theta) + np.cos(2*theta)
y = 2*np.sin(theta) - np.sin(2*theta)

plt.figure()
plt.plot(x, y)
plt.title('Curva deltoide')
plt.xlabel('x'); plt.ylabel('y')
plt.axis('equal')          # eixos na mesma escala
plt.grid(True)

# === Parte (b)  Espiral Galileana r =  ================================
theta_g = np.linspace(0, 10*np.pi, 2000)       # mais pontos para curva suave
r_g = theta_g**2
x_g = r_g * np.cos(theta_g)
y_g = r_g * np.sin(theta_g)

plt.figure()
plt.plot(x_g, y_g)
plt.title('Espiral Galileana  r = ')
plt.xlabel('x'); plt.ylabel('y')
plt.axis('equal')
plt.grid(True)

# === Parte (c)  Funo de Fey ===========================================
theta_f = np.linspace(0, 24*np.pi, 6000)       # curva mais complexa  mais pontos
r_f = np.exp(np.cos(theta_f)) - 2*np.cos(4*theta_f) + np.sin(theta_f/12)**5
x_f = r_f * np.cos(theta_f)
y_f = r_f * np.sin(theta_f)

plt.figure()
plt.plot(x_f, y_f)
plt.title("Funo de Fey")
plt.xlabel('x'); plt.ylabel('y')
plt.axis('equal')
plt.grid(True)

plt.show()
```
